```python
from fillplots import Plotter
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(font_scale=1.25)

plotter = Plotter([[
    (lambda x: 4.8 - x, True),
    (lambda x: 3.5 - 3 * x, ),
    (0, ),
    (lambda x: 0 * x, )
]], xlim=[-1, 6], ylim=[-1, 6])

labels = [
    '$x_1 + x_2 = 4.8$',
    '$-3x_1 - x_2 = -3.5$',
    '$x_1 = 0$',
    '$x_2 = 0$'
]

configs = [
    {'color': 'blue', 'linewidth': 2},
    {'color': 'green', 'linewidth': 2},
    {'color': 'red', 'linewidth': 2, 'linestyle': 'dashed'},
    {'color': 'yellow', 'linewidth': 2, 'linestyle': 'dashed'}
]

plotter.config.line_args = {'color': 'black'}
ineqs = plotter.regions[0].inequalities

for ineq, label, config in zip(ineqs, labels, configs):
    ineq.boundary.config.line_args['label'] = label
    ineq.config.line_args = config

plt.figure(figsize=(5, 5), dpi=150)
plotter.plot()
plotter.ax.legend(loc='best')
plotter.ax.set_xlabel('$x_1$')
plotter.ax.set_ylabel('$x_2$')
```
