```python
from fillplots import Plotter
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set(font_scale=1.4)
```

```python
plotter = Plotter([[
    (lambda x: 4.8  -2 * x / 3, True),
    (lambda x: 4.8 - x, True),
    (lambda x: 3.5 - 3 * x, ),
    (0, ),
    (lambda x: 0 * x, )
]], xlim=[-1, 6], ylim=[-1, 7])

labels = [
    '$2x_1 + 3x_2 = C$',
    '$x_1 + x_2 = 4.8$',
    '$3x_1 + x_2 = 3.5$',
    '$x_1 = 0$',
    '$x_2 = 0$'
]

configs = [
    {'color': '#FF4136', 'linewidth': 2, 'alpha': 0.3},
    {'color': '#39CCCC', 'linewidth': 2},
    {'color': '#0074D9', 'linewidth': 2},
    {'color': '#2ECC40', 'linewidth': 2, 'linestyle': 'dashed'},
    {'color': '#3D9970', 'linewidth': 2, 'linestyle': 'dashed'}
]

plotter.config.line_args = {'color': 'black'}
ineqs = plotter.regions[0].inequalities

for ineq, label, config in zip(ineqs, labels, configs):
    ineq.boundary.config.line_args['label'] = label
    ineq.config.line_args = config

plt.figure(figsize=(5, 5), dpi=150)
plotter.plot()
plotter.ax.legend(loc='best', fontsize=12)
plotter.ax.set_xlabel('$x_1$')
plotter.ax.set_ylabel('$x_2$')
plt.scatter([0, 0, 3.5/3, 4.8], [4.8, 3.5, 0, 0], marker='o', s=50, c='g', zorder=10)
plt.scatter([-0.65, 0], [5.45, 0], marker='o', s=50, c='r', zorder=10)
plt.savefig('../pics/system_1.png', bbox_inches='tight')
```

```python
plotter = Plotter([[
    (lambda x: 4.55  -2 * x / 3, True),
    (lambda x: 4.8 - x, True),
    (lambda x: 3.5 - 3 * x, ),
    (0, ),
    (lambda x: 0 * x, ),
    (lambda x: 0 * x + 4, True)
]], xlim=[-1, 6], ylim=[-1, 7])

labels = [
    '$2x_1 + 3x_2 = C$',
    '$x_1 + x_2 = 4.8$',
    '$3x_1 + x_2 = 3.5$',
    '$x_1 = 0$',
    '$x_2 = 0$',
    '$x_2 = 4$'
]

configs = [
    {'color': '#FF4136', 'linewidth': 2, 'alpha': 0.3},
    {'color': '#39CCCC', 'linewidth': 2},
    {'color': '#0074D9', 'linewidth': 2},
    {'color': '#2ECC40', 'linewidth': 2, 'linestyle': 'dashed'},
    {'color': '#3D9970', 'linewidth': 2, 'linestyle': 'dashed'},
    {'color': '#FF851B', 'linewidth': 2, 'linestyle': 'dashed'}
]

plotter.config.line_args = {'color': 'black'}
ineqs = plotter.regions[0].inequalities

for ineq, label, config in zip(ineqs, labels, configs):
    ineq.boundary.config.line_args['label'] = label
    ineq.config.line_args = config

plt.figure(figsize=(5, 5), dpi=150)
plotter.plot()
plotter.ax.legend(loc='best', fontsize=12.2)
plotter.ax.set_xlabel('$x_1$')
plotter.ax.set_ylabel('$x_2$')
plt.scatter([0, 0, 3.5/3, 4.8, 0.8], [4, 3.5, 0, 0, 4], marker='o', s=50, c='g', zorder=10)
plt.scatter([-0.65, 0], [5.45, 0], marker='o', s=50, c='r', zorder=10)
plt.scatter([0], [4.8], marker='x', s=50, c='r', zorder=10)
plt.savefig('../pics/system_2.png', bbox_inches='tight')
```
